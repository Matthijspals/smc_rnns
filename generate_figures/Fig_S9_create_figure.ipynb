```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

%matplotlib inline
```

```python
# load the saved stats
folder = "../data/hpc/"

# n particles = 1
stats_p1_1 = np.load(folder + "particle1-1.npy")
stats_p1_2 = np.load(folder + "particle1-2.npy")
stats_p1_3 = np.load(folder + "particle1-3.npy")
stats_p1_4 = np.load(folder + "particle1-4.npy")
stats_p1_5 = np.load(folder + "particle1-5.npy")
stats_p1_6 = np.load(folder + "particle1-6.npy")
stats_p1_7 = np.load(folder + "particle1-7.npy")
stats_p1_8 = np.load(folder + "particle1-8.npy")
stats_p1 = [
    stats_p1_1,
    stats_p1_2,
    stats_p1_3,
    stats_p1_4,
    stats_p1_5,
    stats_p1_6,
    stats_p1_7,
    stats_p1_8,
]
n_p1 = 8

# n particles = 10
stats_p10_1 = np.load(folder + "particle10-1.npy")
stats_p10_2 = np.load(folder + "particle10-2.npy")
stats_p10_3 = np.load(folder + "particle10-3.npy")
stats_p10_4 = np.load(folder + "particle10-4.npy")
stats_p10_5 = np.load(folder + "particle10-5.npy")
stats_p10_6 = np.load(folder + "particle10-6.npy")
stats_p10_7 = np.load(folder + "particle10-7.npy")
stats_p10 = [
    stats_p10_1,
    stats_p10_2,
    stats_p10_3,
    stats_p10_4,
    stats_p10_5,
    stats_p10_6,
    stats_p10_7,
]
n_p10 = 7

# n particles = 64
stats_p64_1 = np.load(folder + "particle64-1.npy")
stats_p64_2 = np.load(folder + "particle64-2.npy")  # old run
stats_p64_3 = np.load(folder + "particle64-3.npy")
stats_p64_4 = np.load(folder + "particle64-4.npy")
stats_p64_5 = np.load(folder + "particle64-5.npy")
stats_p64_6 = np.load(folder + "particle64-6.npy")
stats_p64 = [
    stats_p64_1,
    stats_p64_2,
    stats_p64_3,
    stats_p64_4,
    stats_p64_5,
    stats_p64_6,
]
n_p64 = 6

# bootstrap (n particles = 64)
stats_bs1 = np.load(folder + "bs_1.npy")
stats_bs2 = np.load(folder + "bs_2.npy")
stats_bs3 = np.load(folder + "bs_3.npy")
stats_bs = [stats_bs1, stats_bs2, stats_bs3]
n_bs = 3
```

```python
# Aggregate the data for plotting

# bootstrap
r_test_gen_means_bs = [stats_bs[i][0] for i in range(n_bs)]
r_test_gen_isi_means_bs = [stats_bs[i][2] for i in range(n_bs)]
r_test_gen_pair_corrs_bs = [stats_bs[i][4] for i in range(n_bs)]

# particle 1
r_test_gen_means_p1 = [stats_p1[i][0] for i in range(n_p1)]
r_test_gen_isi_means_p1 = [stats_p1[i][2] for i in range(n_p1)]
r_test_gen_pair_corrs_p1 = [stats_p1[i][4] for i in range(n_p1)]

# particle 10
r_test_gen_means_p10 = [stats_p10[i][0] for i in range(n_p10)]
r_test_gen_isi_means_p10 = [stats_p10[i][2] for i in range(n_p10)]
r_test_gen_pair_corrs_p10 = [stats_p10[i][4] for i in range(n_p10)]

# particle 64
r_test_gen_means_p64 = [stats_p64[i][0] for i in range(n_p64)]
r_test_gen_isi_means_p64 = [stats_p64[i][2] for i in range(n_p64)]
r_test_gen_pair_corrs_p64 = [stats_p64[i][4] for i in range(n_p64)]


# do the same for Hellinger distances
D_Hsm_p1 = (
    [stats_p1[i][6] for i in range(n_p1)]
    + [stats_p1[i][7] for i in range(n_p1)]
    + [stats_p1[i][8] for i in range(n_p1)]
)
D_Hsm_p10 = (
    [stats_p10[i][6] for i in range(n_p10)]
    + [stats_p10[i][7] for i in range(n_p10)]
    + [stats_p10[i][8] for i in range(n_p10)]
)
D_Hsm_p64 = (
    [stats_p64[i][6] for i in range(n_p64)]
    + [stats_p64[i][7] for i in range(n_p64)]
    + [stats_p64[i][8] for i in range(n_p64)]
)
D_Hsm_bs = (
    [stats_bs[i][6] for i in range(n_bs)]
    + [stats_bs[i][7] for i in range(n_bs)]
    + [stats_bs[i][8] for i in range(n_bs)]
)


# Train test shif (same in all datasets)
r_test_train_means = stats_bs[0][1]
r_test_train_isi_means = stats_bs[0][3]
r_test_train_pair_corrs = stats_bs[0][5]
```

```python
alpha = 0.5

fig, axes = plt.subplots(1, 2, figsize=(3, 1), dpi=300, width_ratios=[1, 3])
c1 = "teal"
c2 = "darkturquoise"
c3 = "darkslategray"
c4 = "slateblue"
ct = "tomato"

ax = axes[0]

ax.boxplot(
    [D_Hsm_p64],
    positions=[0],
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c2, alpha=alpha), color=c2),
    capprops=dict(color=c2),
    whiskerprops=dict(color=c2),
    medianprops=dict(color=c2),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c2,
        "markeredgecolor": c2,
    },
)

ax.boxplot(
    [D_Hsm_p10],
    positions=[1],
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c1, alpha=alpha), color=c1),
    capprops=dict(color=c1),
    whiskerprops=dict(color=c1),
    medianprops=dict(color=c1),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c1,
        "markeredgecolor": c1,
    },
)

ax.boxplot(
    [D_Hsm_p1],
    positions=[2],
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c3, alpha=alpha), color=c3),
    capprops=dict(color=c3),
    whiskerprops=dict(color=c3),
    medianprops=dict(color=c3),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c3,
        "markeredgecolor": c3,
    },
)

ax.boxplot(
    [D_Hsm_bs],
    positions=[3],
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c4, alpha=alpha), color=c4),
    capprops=dict(color=c4),
    whiskerprops=dict(color=c4),
    medianprops=dict(color=c4),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c4,
        "markeredgecolor": c4,
    },
)

# labels

ax.spines[["bottom"]].set_visible(False)
ax.set_xticks([])
ax.set_ylabel(r"$D_H \downarrow$", rotation=0, labelpad=7)
ax.set_yticks([0.3, 0.5, 0.7, 0.9])
ax.set_yticklabels([".3", ".5", ".7", ".9"])
ax.set_title("LFP")


# Plot 2
pos_64 = [0, 4, 8]
pos_10 = [1, 5, 9]
pos_1 = [2, 6, 10]
pos_bs = [3, 7, 11]

sep = 0.5
pos_64[1] += sep
pos_10[1] += sep
pos_1[1] += sep
pos_bs[1] += sep
pos_64[2] += 2 * sep
pos_10[2] += 2 * sep
pos_1[2] += 2 * sep
pos_bs[2] += 2 * sep

ax = axes[1]

ax.boxplot(
    [r_test_gen_means_bs, r_test_gen_isi_means_bs, r_test_gen_pair_corrs_bs],
    positions=pos_bs,
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c4, alpha=alpha), color=c4),
    capprops=dict(color=c4),
    whiskerprops=dict(color=c4),
    medianprops=dict(color=c4),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c4,
        "markeredgecolor": c4,
    },
    label="bootstrap (k=64)",
)

ax.boxplot(
    [r_test_gen_means_p1, r_test_gen_isi_means_p1, r_test_gen_pair_corrs_p1],
    positions=pos_1,
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c3, alpha=alpha), color=c3),
    capprops=dict(color=c3),
    whiskerprops=dict(color=c3),
    medianprops=dict(color=c3),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c3,
        "markeredgecolor": c3,
    },
    label="k=1",
)

ax.boxplot(
    [r_test_gen_means_p10, r_test_gen_isi_means_p10, r_test_gen_pair_corrs_p10],
    positions=pos_10,
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c1, alpha=alpha), color=c1),
    capprops=dict(color=c1),
    whiskerprops=dict(color=c1),
    medianprops=dict(color=c1),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c1,
        "markeredgecolor": c1,
    },
    label="k=10",
)

ax.boxplot(
    [r_test_gen_means_p64, r_test_gen_isi_means_p64, r_test_gen_pair_corrs_p64],
    positions=pos_64,
    widths=0.6,
    patch_artist=True,
    boxprops=dict(facecolor=mcolors.to_rgba(c2, alpha=alpha), color=c2),
    capprops=dict(color=c2),
    whiskerprops=dict(color=c2),
    medianprops=dict(color=c2),
    flierprops={
        "marker": "o",
        "markersize": 1,
        "markerfacecolor": c2,
        "markeredgecolor": c2,
    },
    label="k=64",
)

# labels
ax.spines[["bottom"]].set_visible(False)
ax.set_ylim(-1, 1)
ax.set_ylabel(r"$r\uparrow$", rotation=0, labelpad=5)
ax.tick_params(axis="x", length=0)
ax.set_xticks([1.5, 5.5, 9.5 + sep * 2])
ax.set_xticklabels([r"$\mu_{rate}$", r"$\mu_{isi}$", "pw corr."])
ax.set_yticks([0, 0.5, 1])
ax.set_yticklabels(["0", ".5", "1"])
ax.set_title("spike stats")

# train test shift
lw = 2
ct = "tomato"
lb = 1
dashes = (2, 0.5)
ax.plot(
    [1.5 - lw, 1.5 + lw],
    [r_test_train_means, r_test_train_means],
    color=ct,
    zorder=-10,
    label=r"train/test",
    lw=lb,
    ls="--",
    dashes=dashes,
)
ax.plot(
    [5.5 + sep - lw, 5.5 + sep + lw],
    [r_test_train_isi_means, r_test_train_isi_means],
    color=ct,
    zorder=-10,
    lw=lb,
    ls="--",
    dashes=dashes,
)
ax.plot(
    [9.5 + sep * 2 - lw, 9.5 + sep * 2 + lw],
    [r_test_train_pair_corrs, r_test_train_pair_corrs],
    color=ct,
    zorder=-10,
    lw=lb,
    ls="--",
    dashes=dashes,
)


# legend
handles, labels = plt.gca().get_legend_handles_labels()
order = [3, 2, 1, 0, 4]
legend = ax.legend(
    [handles[idx] for idx in order],
    [labels[idx] for idx in order],
    loc="upper right",
    bbox_to_anchor=(1.75, 1),
    fontsize=6,
    handlelength=0.55,
    handleheight=0.5,
)
legend_colors = [c2, c1, c3, c4, ct]
for text, color in zip(legend.get_texts(), legend_colors):
    text.set_color(color)

plt.subplots_adjust(wspace=0.5)

plt.savefig("../figures/combined_plots.svg", dpi=300)
plt.savefig("../figures/HPC2_stats.png", dpi=300)
```
